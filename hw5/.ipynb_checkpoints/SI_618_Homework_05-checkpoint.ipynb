````python
custom_stop_words = ['a', 'the', 'and', 'there', 'that', 'this', 'am', 'on',
                     'if', 'it', 'to', 'at' 'a', 'of', 'in', 'out', 'were',
                     'was', 'do', 'did', "don't", "didn't", 'be', 'are', 'is',
                     'being', "it's", 'have', 'had', 's', 'j', 't', 're',
                     'at', 'with', 'just', 'now', "can't", 'can', 'up',
                     'as', 'from', 'thing', 'by', 'so', 'here', 'will', 'for',
                     'll', 've']
````

````python
# define clean function

def clean(x):

    # lower
    x = x.lower()
    # drop http
    x = re.sub("https?:\/\/t.co\/[A-Za-z0-9]*", '', x)

    # stop words
    x = [word for word in x.split() if word not in custom_stop_words]
    x = ' '.join(x)

    # punctuation removal
    x = re.sub(r'[^\w\s]+', ' ', x)
    # num removal
    x = re.sub('[0-9]', '', x)

    # stop words again
    x = [word for word in x.split() if word not in custom_stop_words]
    x = ' '.join(x)

    # removing extra spaces and single letters
    x = re.sub(r'\s+', ' ', x)
    x = re.sub(r'\b\w\b', '', x)

    return x
````

````python
# clean for counting

# first debate
first_debate_biden['cleaned_text'] = first_debate_biden['text'].apply(
    lambda x: clean(x))
first_debate_trump['cleaned_text'] = first_debate_trump['text'].apply(
    lambda x: clean(x))
# second debate
second_debate_biden['cleaned_text'] = second_debate_biden['text'].apply(
    lambda x: clean(x))
second_debate_trump['cleaned_text'] = second_debate_trump['text'].apply(
    lambda x: clean(x))
# biden_town_hall
biden_town_hall_biden['cleaned_text'] = biden_town_hall_biden['text'].apply(
    lambda x: clean(x))
# trump_town_hall
trump_town_hall_trump['cleaned_text'] = trump_town_hall_trump['text'].apply(
    lambda x: clean(x))
````

````python
# define trans_to_turns_df function to get real turn

def trans_to_turns_df(df, biden=False, trump=False):
    blocks = df['speaker'].ne(df['speaker'].shift()).cumsum()
    df_turned = df.groupby([blocks]).agg(
        {'speaker': 'first', 'text': '---------'.join}).reset_index('speaker', drop=True)

    if biden:
        turns_df = len(df_turned[df_turned['speaker'] == 'Joe Biden'])
    if trump:
        turns_df = len(df_turned[df_turned['speaker'] == 'Donald Trump'])

    return turns_df
````

````python
# count number of turns for each debate

# first_debate
display(trans_to_turns_df(first_debate, biden=True, trump=False))
display(trans_to_turns_df(first_debate, biden=False, trump=True))
# second_debate
display(trans_to_turns_df(second_debate, biden=True, trump=False))
display(trans_to_turns_df(second_debate, biden=False, trump=True))
# biden_town_hall
display(trans_to_turns_df(biden_town_hall, biden=True, trump=False))
# trump_town_hall
display(trans_to_turns_df(trump_town_hall, biden=False, trump=True))
````

````python
dict = {'Candidate': ["Trump", "Biden"],
        'First Debate': [trans_to_turns_df(first_debate, biden=False, trump=True), trans_to_turns_df(first_debate, biden=True, trump=False)],
        'Second Debate': [trans_to_turns_df(second_debate, biden=False, trump=True), trans_to_turns_df(second_debate, biden=True, trump=False)],
        'biden_town_hall': ['-', trans_to_turns_df(biden_town_hall, biden=True, trump=False)],
        'trump_town_hall': [trans_to_turns_df(trump_town_hall, biden=False, trump=True), '-'],
        }

turns_count_df = pd.DataFrame(dict).set_index('Candidate')

turns_count_df
````

````python
# load spacy
nlp = spacy.load('en_core_web_sm')

# find noun function


def find_noun(text):
    doc = nlp(text)

    noun_list = []
    for i, sent in enumerate(doc.sents):
        for token in sent:
            if token.pos_ == 'NOUN':
                # print(token.text, "\t", token.pos_)
                noun_list.append(token.lemma_)

    noun_unqi_list = list(set(noun_list))

    return noun_unqi_list
````

````python
# get count of noun

print('The number of different nouns used in first debate by Biden is: ',
      len(find_noun(biden_1)))
print('The number of different nouns used in first debate by trump is: ',
      len(find_noun(trump_1)))
print('The number of different nouns used in second debate by Biden is: ',
      len(find_noun(biden_2)))
print('The number of different nouns used in second debate by trump is: ',
      len(find_noun(trump_2)))
print('The number of different nouns used in biden_town_hall by Biden is: ',
      len(find_noun(biden_hall)))
print('The number of different nouns used in trump_town_hall by trump is: ',
      len(find_noun(trump_hall)))
````

"One fish, two fish, red fish, blue fish"

contains 8 tokens, but 5 types (one, two, red, blue, fish).

Interestingly, the type-token ratio can be used to monitor changes in the use of vocabulary items in children with under-developed vocabulary and/or word finding difficulties and, for example, in adults who have suffered a stroke and who consequently exhibit word retrieval difficulties and naming difficulties. [Ref](https://www.sltinfo.com/type-token-ratio/)
