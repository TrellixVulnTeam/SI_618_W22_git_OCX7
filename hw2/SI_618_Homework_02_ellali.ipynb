```python
# 1. Classify customers into segments based on the total revenue they have generated for the company.
    # low value: less than $1500
    # medium value: between 1500 and 8000 dollars
    # high value: greater than $8000
# Hint: When calculating the segment, we suggest constructing a new dataframe as an intermediary step with the columns: CustomerID, Revenue, Segment.
```

```python
data = pd.merge(invoice_final, purchases_items, how='left', on=['InvoiceNo'])
data
```

```python
Customer_revenue = data.groupby('CustomerID')[['revenue']].sum().sort_values(by=['revenue'],ascending=False)

Customer_revenue['Segment'] = Customer_revenue['revenue']
Customer_revenue.loc[ Customer_revenue['revenue'] < 1500*0.74 , 'Segment'] = 'low value'
Customer_revenue.loc[ (Customer_revenue['revenue'] <= 8000*0.74)&(Customer_revenue['revenue'] >= 1500*0.74), 'Segment'] = 'medium value'
Customer_revenue.loc[ Customer_revenue['revenue'] > 8000*0.74 , 'Segment'] = 'high value'

display(Customer_revenue)
```

```python
# 2. How many customers are in each segment?
print(len(Customer_revenue[Customer_revenue['Segment'] == 'low value']), "customers are in low value.")
print(len(Customer_revenue[Customer_revenue['Segment'] == 'medium value']), "customers are in medium value.")
print(len(Customer_revenue[Customer_revenue['Segment'] == 'high value']), "customers are in high value.")
```
